```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import seaborn as sns

sns.set(style="dark")
```

```python
model_data_set = pickle.load(open('./data/InvertedPendulumSwingupBulletEnv/model_test/reward_data.pkl', 'rb'))
policy_data_set = pickle.load(open('./data/InvertedPendulumSwingupBulletEnv/policy_test/reward_data.pkl', 'rb'))

model_based_data = pickle.load(open('./data/InvertedPendulumSwingupBulletEnv/model_based_results/data.pkl', 'rb'))
model_free_data = pickle.load(open('./data/InvertedPendulumSwingupBulletEnv/model_free_results/data.pkl', 'rb'))
```

```python

plt.figure(figsize=(4,3))

rew = np.mean(model_data_set, axis=0)
std = np.std(model_data_set, axis=0)
vmin = np.max(model_data_set, axis=0)
vmax = np.max(model_data_set, axis=0)

plt.plot(rew[:,0], rew[:,1], label='hybrid learning model only')
plt.fill_between(rew[:,0],
#                 vmin[:,1], vmax[:,1], alpha=0.2)
                    rew[:,1]-std[:,1],rew[:,1]+std[:,1], alpha=0.2)


rew = np.mean(policy_data_set, axis=0)
vmin = np.min(policy_data_set, axis=0)
vmax = np.max(policy_data_set, axis=0)
std = np.std(policy_data_set, axis=0)

plt.plot(rew[:,0], rew[:,1], label='hybrid learning policy only')
plt.fill_between(rew[:,0],
                    rew[:,1]-std[:,1],rew[:,1]+std[:,1], alpha=0.2)

# for data in policy_data_set:
#     plt.plot(data[:,0], data[:,1])
                 
plt.plot(model_based_data['x'], model_based_data['mean'], label='model-based learning')
plt.fill_between(model_based_data['x'], 
                model_based_data['min'], 
                 model_based_data['max'], alpha=0.2)

plt.plot(model_free_data['x'], model_free_data['mean'], label='model-free learning')
plt.fill_between(model_free_data['x'], model_free_data['-std'], model_free_data['+std'], alpha=0.2)

plt.ylabel('episode reward')
plt.xlabel('environment interactions')

plt.legend()

plt.savefig('./component_analysis.pdf')
```
