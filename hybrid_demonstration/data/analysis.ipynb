```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

import glob
```

```python
bc_data = []
for path in glob.glob('behavior_clone/*.pkl'):
    bc_data.append(pickle.load(open(path, 'rb')))
    
hc_data = []
for path in glob.glob('hybrid_control/*.pkl'):
    hc_data.append(pickle.load(open(path, 'rb')))
```

```python
x = [i for i in range(len(bc_data[0]))]
bc_mean = np.mean(bc_data, axis=0)
bc_std = np.std(bc_data, axis=0)

hc_mean = np.mean(hc_data, axis=0)
hc_std = np.std(hc_data, axis=0)

plt.plot(x,bc_mean)
plt.fill_between(x, bc_mean-bc_std, bc_mean+bc_std , alpha=0.2)

plt.plot(np.mean(hc_data, axis=0))
plt.fill_between(x, hc_mean-hc_std, hc_mean+hc_std , alpha=0.2)
```

```python
plt.plot(np.stack(bc_data).T, 'k')
plt.plot(np.stack(hc_data).T, 'r')
```
